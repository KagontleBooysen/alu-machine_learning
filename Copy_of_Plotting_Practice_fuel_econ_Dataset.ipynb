```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your data into a DataFrame (replace 'data.csv' with your dataset's file name)
data = pd.read_csv("/content/drive/MyDrive/fuel_econ.csv")

# Compute the correlation matrix
correlation_matrix = data.corr()

# Filter the correlation matrix to show only positive correlations close to 1
threshold = 0.9  # Adjust the threshold as needed
positive_correlations = correlation_matrix[correlation_matrix > threshold]

# Set the diagonal elements to NaN to avoid self-correlations
np.fill_diagonal(positive_correlations.values, np.nan)

# Create a heatmap for positive correlations
plt.figure(figsize=(10, 8))
sns.heatmap(positive_correlations, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Positive Correlations Close to 1")
plt.show()
```

```python
# This correlation coefficient of 0.86 between "Co2" and "displacement" indicates a strong positive linear correlation between these two variables.
# The positive sign of the correlation coefficient (0.86) indicates a positive correlation, meaning that as one variable (CO2 emissions) increases, the other variable (displacement) tends to increase as well. In other words, when CO2 emissions go up, there is a tendency for displacement to also increase.
```

```python
import pandas as pd

# Load your data into a DataFrame (replace 'data.csv' with your dataset's file name)
data = pd.read_csv("/content/drive/MyDrive/fuel_econ.csv")

# Extract the 'co2', 'displ', and 'cylinders' columns
co2_emissions = data['co2']
displacement = data['displ']
cylinders = data['cylinders']

# Calculate the correlation matrix
correlation_matrix = data[['co2', 'displ', 'cylinders']].corr()

print(correlation_matrix)
```

```python
#'co2' has a strong positive correlation with 'displ' (approximately 0.76) and 'cylinders' (approximately 0.74).
#'displ' has a strong positive correlation with 'cylinders' (approximately 0.92).
# As 'co2' emissions increase, 'displacement' and 'cylinder' count tend to increase.
# As 'displacement' increases, the 'cylinder' count tends to increase.
```
